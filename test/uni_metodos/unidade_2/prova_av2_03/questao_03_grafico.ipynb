```python
import numpy as np
import matplotlib.pyplot as plt

# --- Questão 3: Altura e Peso de Funcionários ---

# Dados fornecidos na tabela da Questão 3
# Altura (cm) - variável independente (x)
alturas_cm_originais = np.array([183, 173, 168, 188, 158, 163, 193, 163, 178])
# Peso (kg) - variável dependente (y)
pesos_kg_originais = np.array([79,  69,  70,  81,  61,  63,  79,  71,  73])

# Passo 1: Ajustar uma reta que descreve o comportamento do peso em função da altura.
# O modelo é: peso = m * altura + c
coeficientes = np.polyfit(alturas_cm_originais, pesos_kg_originais, 1)
m_slope = coeficientes[0]
c_intercept = coeficientes[1]

print("--- Análise da Relação entre Altura e Peso ---")
print(f"A equação da reta de regressão ajustada é:")
print(f"Peso (kg) = {m_slope:.4f} * Altura (cm) + ({c_intercept:.4f})")
print("-" * 50)

# Passo 2: Com base nessas informações, responder:

# a) Estime o peso de um funcionário com 175 cm de altura
altura_especifica_a = 175
peso_estimado_a = m_slope * altura_especifica_a + c_intercept

print(f"a) Estimativa do peso para um funcionário com {altura_especifica_a} cm de altura:")
print(f"   Peso estimado: {peso_estimado_a:.2f} kg")
print("-" * 50)

# b) Estime a altura de um funcionário com 80 kg
peso_especifico_b = 80
altura_estimada_b = (peso_especifico_b - c_intercept) / m_slope

print(f"b) Estimativa da altura para um funcionário com {peso_especifico_b} kg:")
print(f"   Altura estimada: {altura_estimada_b:.2f} cm")
print("-" * 50)

# --- Geração do Gráfico ---
plt.figure(figsize=(10, 6))

# Plotar os dados originais
plt.scatter(alturas_cm_originais, pesos_kg_originais, color='blue', label='Dados Originais (Observados)', zorder=5)

# Criar uma sequência de valores de x (altura) para plotar a reta ajustada
# Usamos o mínimo e máximo das alturas originais para definir o intervalo da reta
x_para_reta = np.array([min(alturas_cm_originais) - 5, max(alturas_cm_originais) + 5]) # Estende um pouco as bordas
y_reta_ajustada = m_slope * x_para_reta + c_intercept

# Plotar a reta de regressão ajustada
plt.plot(x_para_reta, y_reta_ajustada, color='red', linestyle='--', label=f'Reta Ajustada: P={m_slope:.2f}A+({c_intercept:.2f})')

# Plotar os pontos estimados
# Ponto (a): (175 cm, peso_estimado_a)
plt.scatter([altura_especifica_a], [peso_estimado_a], color='green', s=100, edgecolor='black', label=f'Estimativa (a): ({altura_especifica_a}cm, {peso_estimado_a:.2f}kg)', zorder=6)
# Ponto (b): (altura_estimada_b, 80 kg)
plt.scatter([altura_estimada_b], [peso_especifico_b], color='purple', s=100, edgecolor='black', label=f'Estimativa (b): ({altura_estimada_b:.2f}cm, {peso_especifico_b}kg)', zorder=6)


# Configurações do gráfico
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Regressão Linear: Peso em Função da Altura de Funcionários')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(min(alturas_cm_originais)-10, color='black', linewidth=0.5) # Ajustar o início do eixo x se necessário

# Mostrar o gráfico
plt.show()
```
