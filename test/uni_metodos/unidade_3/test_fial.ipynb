```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

def gerar_grafico_comparativo(df: pd.DataFrame):
    """
    Gera um gráfico de linhas comparativo a partir de um DataFrame.

    Cada coluna do DataFrame é plotada como uma linha separada,
    com o índice do DataFrame servindo como eixo X.

    Args:
        df (pd.DataFrame): O DataFrame contendo os dados a serem plotados.
                           As colunas devem ser as séries de dados (ex: 'fun_euler')
                           e o índice deve ser as iterações.
    """
    # Verificação para garantir que a entrada é um DataFrame e não está vazio
    if not isinstance(df, pd.DataFrame) or df.empty:
        print("Erro: A entrada deve ser um DataFrame do Pandas não vazio.")
        return

    # Define um estilo para o gráfico e cria a figura e os eixos
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(14, 8))

    # Itera sobre cada coluna no DataFrame
    for coluna in df.columns:
        # Plota os dados da coluna atual
        # df.index é usado para o eixo X (Iterações 0, 1, 2, ...)
        # df[coluna] é usado para o eixo Y (os valores calculados)
        ax.plot(df.index, df[coluna], marker='o', linestyle='-', label=coluna)

    # --- Adiciona títulos e legendas para maior clareza ---
    ax.set_title('Gráfico Comparativo de Métodos Numéricos', fontsize=18, pad=20)
    ax.set_xlabel('Iteração (Passo)', fontsize=14)
    ax.set_ylabel('Valor Calculado / Erro', fontsize=14)
    ax.legend(title='Métodos', fontsize=12)

    # Garante que o layout do gráfico fique bem ajustado
    plt.tight_layout()

    # Exibe o gráfico gerado
    plt.show()
```

```python
gerar_grafico_comparativo(dataframe_final)
```
